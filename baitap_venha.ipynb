```python
import copy
from collections import deque

board = [[2,8,3],[1,6,4],[7,0,5]]
goal = [[1,2,3],[8,0,4],[7,6,5]]

move = {
    (-1,0): "UP",
    (1,0): "DOWN",
    (0,-1): "LEFT",
    (0,1): "RIGHT"
}


def goal_test(state):
    for i in range(len(goal)):
        for j in range(len(goal[0])):
            if state[i][j] != goal[i][j]:
                return False
    return True


def find_zero(state):
    for i in range(len(state)):
        for j in range(len(state[0])):
            if state[i][j] == 0:
                return i,j


class node:

    def __init__(self,state_cur,parent_node,action,path_cost):
        self.state_cur = state_cur
        self.parent_node = parent_node
        self.action = action
        self.path_cost = path_cost

    def getPath_cost(self):
        return self.path_cost

    def __str__(self):

        parent_state = None

        if self.parent_node is not None:
            parent_state = self.parent_node.state_cur

        return (
            f"state_cur: {self.state_cur}\n"
            f"parent_state: {parent_state}\n"
            f"action: {self.action}\n"
            f"path_cost: {self.path_cost}\n"
            f"-------------------------------------------"
        )


def trace_path(goal):

    path = []

    cur = goal

    while cur is not None:
        path.append(cur)
        cur = cur.parent_node

    path.reverse()

    return path


def in_path(cur_node,state):

    while cur_node is not None:

        if cur_node.state_cur == state:
            return True

        cur_node = cur_node.parent_node

    return False


def Depth_Limited_Search(problem,gioi_han):

    frontier = []

    direction = [
        (-1,0),
        (1,0),
        (0,-1),
        (0,1)
    ]

    root = node(
        problem,
        None,
        None,
        0
    )

    frontier.append(root)

    ket_qua = "failure"

    while frontier:

        cur_node = frontier.pop()

        if goal_test(
            cur_node.state_cur
        ):
            return cur_node


        if cur_node.path_cost >= gioi_han:

            ket_qua = "cutoff"

        else:

            x,y = find_zero(
                cur_node.state_cur
            )

            for dx,dy in direction:

                nx = x + dx
                ny = y + dy

                if(
                    0<=nx<len(board)
                    and
                    0<=ny<len(board[0])
                ):

                    new_state = copy.deepcopy(
                        cur_node.state_cur
                    )

                    new_state[x][y],new_state[nx][ny] = \
                    new_state[nx][ny],new_state[x][y]


                    if not in_path(
                        cur_node,
                        new_state
                    ):

                        child = node(
                            new_state,
                            cur_node,
                            move[(dx,dy)],
                            cur_node.path_cost+1
                        )

                        frontier.append(
                            child
                        )

    return ket_qua



def Iterative_Deepening_Search(problem):

    gioi_han = 0

    while True:

        print(
            f"Thử độ sâu = {gioi_han}"
        )

        ket_qua = Depth_Limited_Search(
            problem,
            gioi_han
        )

        if ket_qua == "cutoff":

            gioi_han += 1

        elif ket_qua == "failure":

            return None

        else:

            return ket_qua



result = Iterative_Deepening_Search(
    board
)

if result:

    path = trace_path(
        result
    )

    for state in path:
        print(state)

else:
    print(
        "No Solution"
    )
```

```python
import sys
import copy
import tkinter as tk

sys.setrecursionlimit(100000)

board = [[2,8,3],[1,6,4],[7,0,5]]
goal = [[1,2,3],[8,0,4],[7,6,5]]

move = {
    (-1,0): "UP",
    (1,0): "DOWN",
    (0,-1): "LEFT",
    (0,1): "RIGHT"
}


def goal_test(state):
    for i in range(len(goal)):
        for j in range(len(goal[0])):
            if state[i][j] != goal[i][j]:
                return False
    return True


def find_zero(state):

    for i in range(len(state)):
        for j in range(len(state[0])):
            if state[i][j]==0:
                return i,j


class node:

    def __init__(
        self,
        state_cur,
        parent_node,
        action,
        path_cost
    ):

        self.state_cur=state_cur
        self.parent_node=parent_node
        self.action=action
        self.path_cost=path_cost


    def getPath_cost(self):
        return self.path_cost


    def __str__(self):

        parent_state=None

        if self.parent_node is not None:
            parent_state=self.parent_node.state_cur

        return(
            f"state_cur:{self.state_cur}\n"
            f"parent_state:{parent_state}\n"
            f"action:{self.action}\n"
            f"path_cost:{self.path_cost}\n"
        )


def state_to_tuple(state):

    return tuple(
        tuple(row)
        for row in state
    )


def Depth_First_Search(problem):

    stack=[]

    explored=set()

    direction=[
        (0,1),
        (0,-1),
        (1,0),
        (-1,0)
    ]

    root=node(
        problem,
        None,
        None,
        0
    )

    explored.add(
        state_to_tuple(
            problem
        )
    )

    stack.append(
        root
    )

    while stack:

        cur_node=stack.pop()

        if goal_test(
            cur_node.state_cur
        ):
            return cur_node


        x,y=find_zero(
            cur_node.state_cur
        )


        for dx,dy in direction:

            nx=x+dx
            ny=y+dy

            if(
                0<=nx<3
                and
                0<=ny<3
            ):

                new_state=copy.deepcopy(
                    cur_node.state_cur
                )

                new_state[x][y],new_state[nx][ny]=\
                new_state[nx][ny],new_state[x][y]

                key=state_to_tuple(
                    new_state
                )

                if key not in explored:

                    explored.add(
                        key
                    )

                    child=node(
                        new_state,
                        cur_node,
                        move[(dx,dy)],
                        cur_node.path_cost+1
                    )

                    stack.append(
                        child
                    )

    return None


def trace_path(goal):

    path=[]

    cur=goal

    while cur is not None:

        path.append(
            cur
        )

        cur=cur.parent_node

    path.reverse()

    return path


class PuzzleGUI:

    def __init__(self,root):

        self.root=root

        root.title(
            "DFS Visualizer"
        )

        self.cells=[]

        frame=tk.Frame(root)
        frame.pack(
            pady=20
        )

        for i in range(3):

            row=[]

            for j in range(3):

                label=tk.Label(
                    frame,
                    width=4,
                    height=2,
                    font=("Arial",22),
                    relief="solid"
                )

                label.grid(
                    row=i,
                    column=j,
                    padx=3,
                    pady=3
                )

                row.append(
                    label
                )

            self.cells.append(
                row
            )


        self.info=tk.Label(
            root,
            font=("Arial",13)
        )

        self.info.pack()


        button=tk.Button(
            root,
            text="Start DFS",
            command=self.start
        )

        button.pack(
            pady=10
        )


    def draw(self,state):

        for i in range(3):

            for j in range(3):

                value=state[i][j]

                if value==0:

                    self.cells[i][j][
                        "text"
                    ]=""

                else:

                    self.cells[i][j][
                        "text"
                    ]=str(value)



    def start(self):

        result=Depth_First_Search(
            board
        )

        if result is None:

            self.info[
                "text"
            ]="No Solution"

            return


        self.path=trace_path(
            result
        )

        self.index=0

        self.animate()



    def animate(self):

        if self.index>=len(
            self.path
        ):

            self.info[
                "text"
            ]+="\nGoal Found"

            return


        current=(
            self.path[
                self.index
            ]
        )

        self.draw(
            current.state_cur
        )

        self.info[
            "text"
        ]=(
            f"Action: {current.action}\n"
            f"Path Cost: {current.path_cost}"
        )

        self.index+=1

        self.root.after(
            1000,
            self.animate
        )


root=tk.Tk()

app=PuzzleGUI(
    root
)

root.mainloop()
```
