```python
import copy
from collections import deque
board = [[2,8,3],[1,6,4],[7,0,5]]
goal = [[1,2,3],[8,0,4],[7,6,5]]
move = {
    (-1,0): "UP",
    (1,0): "DOWN",
    (0,-1): "LEFT",
    (0,1): "RIGHT"
}

def goal_test(state):
    for i in range(len(goal)):
        for j in range(len(goal[0])):
            if state[i][j] != goal[i][j]:
                return False
    return True
def find_zero(state):
    for i in range(len(state)):
        for j in range(len(state[0])):
            if state[i][j] == 0:
                return i,j
class node:
    def __init__(self,state_cur,parent_node,action,path_cost):
        self.state_cur = state_cur
        self.parent_node = parent_node
        self.action = action
        self.path_cost = path_cost
    def getPath_cost(self):
        return self.path_cost
    def __str__(self):

        parent_state = None

        if self.parent_node is not None:
            parent_state = self.parent_node.state_cur

        return (
            f"state_cur: {self.state_cur}\n"
            f"parent_state: {parent_state}\n"
            f"action: {self.action}\n"
            f"path_cost: {self.path_cost}\n"
            f"-------------------------------------------"
        )
def trace_path(goal):
    path = []
    cur = goal
    while cur is not None:
        path.append(cur)
        cur = cur.parent_node
    path.reverse()
    return path
def Breadth_First_Search(problem):
    explored = set()
    frontier = deque()
    direction = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    root = node(problem,None,None,0)
    explored.add(str(problem))
    if goal_test(problem):
            return root
    frontier.append(root)
    while frontier:
        cur_node = frontier.popleft()
        x,y = find_zero(cur_node.state_cur)
        for dx,dy in direction:
            nx = x + dx
            ny = y + dy
            if 0 <= nx < len(board) and 0 <= ny < len(board[0]):
                new_state = copy.deepcopy(cur_node.state_cur)
                new_state[x][y],new_state[nx][ny] = new_state[nx][ny],new_state[x][y]
                if str(new_state) not in explored:
                    child = node(new_state,cur_node,move[(dx,dy)],cur_node.path_cost+1)
                    explored.add(str(child.state_cur))
                    if goal_test(child.state_cur):
                          return child
                    frontier.append(child)
    return None
import tkinter as tk
from tkinter import ttk


class PuzzleGUI:

    def __init__(self,root):

        self.root=root

        root.title(
            "8 Puzzle BFS"
        )

        top=tk.Frame(
            root
        )

        top.pack(
            pady=10
        )


        tk.Label(
            top,
            text="Algorithm:"
        ).pack(
            side="left"
        )


        self.combo=ttk.Combobox(
            top,
            values=[
                "BFS",
                "DFS",
                "IDS"
            ],
            state="readonly"
        )

        self.combo.current(
            0
        )

        self.combo.pack(
            side="left"
        )


        tk.Button(
            top,
            text="Solve",
            command=self.start
        ).pack(
            side="left"
        )


        frame=tk.Frame(
            root
        )

        frame.pack()


        self.start_board=\
        self.create_board(
            frame,
            0,
            "START"
        )


        self.current_board=\
        self.create_board(
            frame,
            1,
            "CURRENT"
        )


        self.goal_board=\
        self.create_board(
            frame,
            2,
            "GOAL"
        )


        self.draw(
            self.start_board,
            board
        )

        self.draw(
            self.goal_board,
            goal
        )


        self.info=tk.Label(
            root,
            font=(
                "Arial",
                12
            )
        )

        self.info.pack(
            pady=10
        )


    def create_board(
            self,
            parent,
            col,
            title
    ):

        outer=tk.Frame(
            parent
        )

        outer.grid(
            row=0,
            column=col,
            padx=20
        )


        tk.Label(
            outer,
            text=title
        ).pack()


        grid=tk.Frame(
            outer
        )

        grid.pack()

        cells=[]

        for i in range(3):

            row=[]

            for j in range(3):

                lbl=tk.Label(
                    grid,
                    width=4,
                    height=2,
                    font=(
                        "Arial",
                        20
                    ),
                    relief="solid"
                )

                lbl.grid(
                    row=i,
                    column=j
                )

                row.append(
                    lbl
                )

            cells.append(
                row
            )

        return cells


    def draw(
            self,
            cells,
            state
    ):

        for i in range(3):
            for j in range(3):

                value=state[i][j]

                cells[i][j]["text"]=(
                    ""
                    if value==0
                    else value
                )


    def start(self):

        algo=self.combo.get()

        if algo=="BFS":
            result=Breadth_First_Search(board)

        elif algo=="DFS":
            result=Depth_First_Search(board)

        elif algo=="BFS":
            result=Breadth_First_Search(board)

        if result is None:
            self.info["text"]="No solution"
            return

        self.path=trace_path(result)

        self.index=0

        self.animate()


    def animate(self):

        if self.index>=len(
                self.path
        ):

            return


        cur=\
        self.path[
            self.index
        ]


        self.draw(
            self.current_board,
            cur.state_cur
        )


        self.info[
            "text"
        ]=(
            f"Action:{cur.action}\n"
            f"Cost:{cur.path_cost}"
        )


        self.index+=1

        self.root.after(
            700,
            self.animate
        )


root=tk.Tk()

PuzzleGUI(
    root
)

root.mainloop()
result = Breadth_First_Search(board)
path = trace_path(result)
for state in path:
    print(state)
```

```python
from collections import deque
import copy
moves = {
    (-1,0):"UP",
    (1,0):"DOWN",
    (0,-1):"LEFT",
    (0,1):"RIGHT"
}
class Node:
    def __init__(self,state,parent_node,action,path_cost):
        self.state = state
        self.parent_node = parent_node
        self.action = action
        self.path_cost = path_cost
    def __str__(self):
        return (f"state: {self.state}\n"
                f"parent_node: {self.parent_node is not None}\n"
                f"action: {self.action}\n"
                f"path_cost: {self.path_cost}")
def input_problem():
    root = []
    rows = -1
    cols = -1
    while True:
        try:
            rows = int(input("nhập số hàng: "))
            cols = int(input("nhập số cột: "))
            break
        except:
            print("yêu cầu nhập lại")
    for i in range(rows):
        row = []
        for j in range(cols):
           row.append((input("nhập phần tử tại ({},{}): ".format(i,j))))
        root.append(row)
    return root
def test_goal(cur_state, goal_pos):
    ax, ay = get_position_start(cur_state)
    return (ax, ay) == goal_pos
def get_position_end(goal_state):
    for i in range(len(goal_state)):
        for j in range(len(goal_state[0])):
            if goal_state[i][j] == "B":
                return (i,j)
def get_position_start(problem):
    for i in range(len(problem)):
        for j in range(len(problem[0])):
            if problem[i][j] == "A":
                return (i,j)
def trace_path(goal_node):
    path = []
    temp = goal_node
    while temp is not None:
        path.append(temp)
        temp = temp.parent_node
    path.reverse()
    return path
def breadth_first_search(problem):
    frontier = deque()
    explored = set()
    direction = [(-1,0),(1,0),(0,-1),(0,1)]
    root = Node(problem,None,None,0)
    frontier.append(root)
    while frontier:
        cur_node = frontier.popleft()

        if test_goal(cur_node.state,goal_pos):
            return cur_node
        x,y = get_position_start(cur_node.state)
        rows = len(cur_node.state)
        cols = len(cur_node.state[0])
        explored.add(str(cur_node.state))
        for i,j in direction:
            nx = x+i
            ny = y+j
            new_state = copy.deepcopy(cur_node.state)
            if 0 <= nx < rows and 0 <= ny < cols and new_state[nx][ny] != "#"  :
                new_state[x][y],new_state[nx][ny] = new_state[nx][ny],new_state[x][y]
                if str(new_state) not in explored:
                    new_node = Node(new_state,cur_node,moves[(i,j)],cur_node.path_cost+1)
                    frontier.append(new_node)
                    explored.add(str(new_state))
    return None
problem = [
["A",".","."],
["#","#","."],
[".",".","B"]
]
goal_pos = get_position_end(problem)
goal = breadth_first_search(problem)
path = trace_path(goal)
for state in path:
    print(state)
```
